## Data Analysis & Visualization 

### 1. Countplot displaying the transition of songs from 1985 -2023 

```python
# visualization of count plot showing transition of songs wrt their count 
sns.set(rc ={"font.size":13,"axes.titlesize":13,"axes.labelsize":5,'figure.figsize' : (15,5.65)})
palette = ["#F72585", "#7209B7", "#3A0CA3", "#4361EE", "#4CC9F0"]
ax= sns.countplot(x = 'year', data =df, palette = palette)
plt.xlabel('year', fontsize=20);
plt.ylabel('count', fontsize=20);
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df
```

####  Top Artists

```python
# ranking the artists based on their popularity by creating df_top_artist
df_top_artist = df.groupby(['artist'], as_index = False)['popularity'].sum().sort_values(by = 'popularity', ascending  = False)
df_top_artist
```

```python
# Sorting the song titles in Descending order
df['title'].sort_values(ascending = False)
```

```python
# grouping the data by 'title' and sorting it by sum of 'energy' in descending order
df_title = df.groupby(['title'])['energy'].sum().sort_values(ascending = False)
df_title
```

### 2. Lineplot of artists by popularity

```python
# analysing the aggregate components such as mean,max,min,count,etc of df_top_artist
df_top_artist.describe()
```

```python
# visualization of line plot showing artists by their popularity 
sns.set(rc ={"font.size":8,"axes.titlesize":8,"axes.labelsize":5,'figure.figsize' : (5,7)})
ar_popu = sns.lineplot(x = 'artist', y = 'popularity', data = df_top_artist)
ar_popu
plt.xlabel('artist', fontsize=15);
plt.ylabel('popularity', fontsize=15);
```

### 3. Violin plot of dB levels

```python
# analysing the aggregate components such as mean,max,min,count,etc of 'dB' column
df['dB'].describe()
```

```python
# visualization of violin plot showing dB levels by their count 
sns.set(rc ={"font.size":8,"axes.titlesize":8,"axes.labelsize":5,'figure.figsize' : (7,8)})
vp = sns.violinplot(data = df, x='dB', palette = "Blues")

plt.xlabel('dB', fontsize=15);

for values in vp.containers:
    vp.violin_label(values)
```

### 4. Countplot plot of bpm 

```python
# analysing the aggregate components such as mean,max,min,count,etc of 'bpm' column
df['bpm'].describe()
```

```python
# visualization of count plot showing 'bpm'(beats per minute) levels by their count 
sns.set(rc ={"font.size":8,"axes.titlesize":8,"axes.labelsize":5,'figure.figsize':(100,48)})
cx = sns.countplot(data = df, y= 'bpm', palette = "Greens_d", order  = df.bpm.value_counts().iloc[:].index)
plt.xlabel('count', fontsize=150);
plt.ylabel('bpm', fontsize=150);
# for bars in cx.containers:
#         cx.bar_label(bars)
```

### 5. Barplot presenting yearwise changes in the liveness of the songs

```python
# analysing the aggregate components such as mean,max,min,count,etc of 'year' & 'liveness' column
df[['year','liveness']].describe(include = 'all')
```

```python
# visualization of bar plot showing liveness of available songs in consecutive years 
year_liveness = df.groupby(['year'], as_index=False)['liveness'].sum().sort_values(by = 'liveness', ascending = True)
sns.set(rc ={"font.size":13,"axes.titlesize":8,"axes.labelsize":5,'figure.figsize':(15,6)})
bp = sns.barplot(data = year_liveness, x = 'year' , y = 'liveness')
plt.xlabel('year', fontsize=20);
plt.ylabel('liveness', fontsize=20);
for values in bp.containers :
    bp.bar_label(values)
```
